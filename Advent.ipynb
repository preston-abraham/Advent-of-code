```python
f = open('day1.txt', 'r')

# Create a list to store the inventory of each elf
elf_inventories = []

# Parse the file and store the inventory of each elf
inventory = []
for line in f:
    if line == '\n':
        elf_inventories.append(inventory)
        inventory = []
    else:
        inventory.append(int(line))

# Calculate the total calories carried by each elf
elf_totals = []
for elf_inventory in elf_inventories:
    elf_totals.append(sum(elf_inventory))

# Sort the elf totals in descending order
sorted_elf_totals = sorted(elf_totals, reverse=True)

# Calculate the total calories carried by the top 3 elves
top_3_total = sum(sorted_elf_totals[:3])

# Print the result
print(f'The top 3 elves are carrying a total of {top_3_total} Calories.')
```

```python
#Read the file
file = open('Day2.txt', 'r')

#Create variables to store the scores
total_score = 0

#Loop through each line in the file
for line in file:
  #Split the line into two parts
    opponent_choice, player_choice = line.strip().split()

  #Calculate the score for the round
    if (opponent_choice == 'A' and player_choice == 'X') or (opponent_choice == 'B' and player_choice == 'Y') or (opponent_choice == 'C' and player_choice == 'Z'):
        round_score = 3
    elif (opponent_choice == 'A' and player_choice == 'Z') or (opponent_choice == 'B' and player_choice == 'X') or (opponent_choice == 'C' and player_choice == 'Y'):
        round_score = 0
    else:
        round_score = 6

  #Calculate the score for the shape selected
    if player_choice == 'X':
        shape_score = 1
    elif player_choice == 'Y':
        shape_score = 2
    elif player_choice == 'Z':
        shape_score = 3

  #Calculate the total score for the round
    total_score += round_score + shape_score

#Print the total score
print(total_score)
```

```python
# Solution
def calculateScore(file):
    with open(file, 'r') as f:
        totalScore = 0
        for line in f:
            opponent, outcome = line.strip().split(' ')
            if opponent == 'A':
                if outcome == 'X':
                    totalScore += 3
                elif outcome == 'Y':
                    totalScore += 4
                elif outcome == 'Z':
                    totalScore += 8
            elif opponent == 'B':
                if outcome == 'X':
                    totalScore += 1
                elif outcome == 'Y':
                    totalScore += 5
                elif outcome == 'Z':
                    totalScore += 9
            elif opponent == 'C':
                if outcome == 'X':
                    totalScore += 2
                elif outcome == 'Y':
                    totalScore += 6
                elif outcome == 'Z':
                    totalScore += 7
        return totalScore

print(calculateScore('day2.txt'))
```

```python
# open the day3.txt file
with open('day3.txt', 'r') as f:
  # read the contents of the file
  contents = f.read()

# split the contents of the file into a list of strings
contents = contents.split()

# create an empty list to store the priorities
priorities = []

# iterate through the list of strings
for line in contents:
  # the length of the line divided by 2 will give the number of items in each compartment
  half_length = int(len(line)/2)
  # the first half of the characters represent items in the first compartment
  first_compartment = line[:half_length]
  # the second half of the characters represent items in the second compartment
  second_compartment = line[half_length:]
  # create two empty sets to store the items in each compartment
  items_1 = set()
  items_2 = set()
  # iterate through the first compartment
  for i in first_compartment:
    # add the item to the set
    items_1.add(i)
  # iterate through the second compartment
  for j in second_compartment:
    # add the item to the set
    items_2.add(j)
  # find the items that appear in both compartments
  common_items = items_1.intersection(items_2)
  # iterate through the common items
  for item in common_items:
    # if the item is lowercase, assign it a priority between 1 and 26
    if item.islower():
      priority = ord(item) - 96
    # if the item is uppercase, assign it a priority between 27 and 52
    elif item.isupper():
      priority = ord(item) - 38
    # add the priority to the list
    priorities.append(priority)

# calculate the sum of the priorities
sum_priorities = sum(priorities)

# print the result
print("The sum of the priorities of the item types that appear in both compartments of each rucksack is {}".format(sum_priorities))
```

```python
# Read the text file
with open('day4.txt', 'r') as f:
    data = f.readlines()

# Initialize the counter
counter = 0

# Loop through the lines in the text file
for line in data:
    # Split the line into two parts
    part1, part2 = line.split(',')
    # Extract the start and end numbers
    start1, end1 = int(part1.split('-')[0]), int(part1.split('-')[1])
    start2, end2 = int(part2.split('-')[0]), int(part2.split('-')[1])
    # Check if one range contains the other
    if (start1 <= start2 and end1 >= end2) or (start2 <= start1 and end2 >= end1):
        # Increment the counter
        counter += 1

# Print the answer
print(f'In {counter} assignment pairs, one range fully contains the other.')
```

```python
#Solution

# Read the text file
f = open("day4.txt", "r")

# Initialize a counter
overlap_count = 0

# Loop through the file
for line in f:
    # Split the line into two range strings
    ranges = line.split(',')
    # Split each range into two numbers
    range1 = ranges[0].split('-')
    range2 = ranges[1].split('-')
    # Check if the ranges overlap
    if int(range2[0]) <= int(range1[1]) and int(range1[0]) <= int(range2[1]):
        overlap_count += 1

# Print the result
print("The number of assignment pairs with overlapping ranges is:", overlap_count)
```

```python
# Read the stacks from the file
stacks = []
with open('day5_stacks.txt') as f:
    for line in f:
        stacks.append(line.split())

# Read the moves from the file
moves = []
with open('day5.txt') as f:
    for line in f:
        moves.append(line.strip().split())

# Perform the moves
for move in moves[:]:
    num_crates = int(move[1])
    from_stack = int(move[3])-1
    to_stack = int(move[5])-1
    for _ in range(num_crates):
        crate = stacks[from_stack].pop(0)
        stacks[to_stack].insert(0, crate)
# Print the top crate of each stack
for i, stack in enumerate(stacks):
    print(f'Stack {i+1}: {stack[0]}')
```

```python
# Open the files

stacks = []
with open('day5_stacks.txt') as f:
    for line in f:
        stacks.append(line.split())

# Read the moves from the file
moves = []
with open('day5.txt') as f:
    for line in f:
        moves.append(line.strip().split())

# Perform the moves
for move in moves:
    num_crates = int(move[1])
    stack_from = int(move[3])-1
    stack_to = int(move[5])-1
    crates_transferred = stacks[stack_from][:num_crates]
    del stacks[stack_from][:num_crates]
    stacks[stack_to] = crates_transferred + stacks[stack_to]

# Print out the top crate from each stack
for i in range(len(stacks)):
    print("Stack " + str(i) + ": " + stacks[i][0])
```

```python
grid = pd.DataFrame([list(i.strip()) for i in open('day8.txt').readlines()])
score_grid = grid.copy()
visible_trees = 0
for i in range(99):
    for j in range(99):
        visible = 4
        height = grid.at[i,j]
        score = [0,0,0,0]
        #Left
        for k in range(1,i+1):
            score[0] += 1
            if grid.at[i-k,j] >= height:
                visible -= 1
                break
                
        #Right
        for k in range(i+1,99):
            score[1] += 1
            if grid.at[k,j] >= height:
                visible -= 1
                break
        #Top
        for k in range(1,j+1):
            score[2] += 1
            if grid.at[i,j-k] >= height:
                visible -= 1
                break
        #Bottom
        for k in range(j+1,99):
            score[3] += 1
            if grid.at[i,k] >= height:
                visible -= 1
                break
                
        if visible > 0:
            visible_trees+=1
            
        score_grid.at[i,j] = np.product(score)
        
print(visible_trees)
print(max(score_grid.max(axis=0)))
```

```python
moves = [i.strip() for i in open('day9.txt').readlines()]
rope_length = 10
rope_pos = [[0,0] for i in range(rope_length)]
visited = []
for move in moves:
    num_steps = int(move.split()[1])
    for step in range(num_steps):
        if move.split()[0] == 'L':
            rope_pos[0][0]-=1
        elif move.split()[0] == 'R':
            rope_pos[0][0]+=1
        elif move.split()[0] == 'U':
            rope_pos[0][1]+=1
        elif move.split()[0] == 'D':
            rope_pos[0][1]-=1
        
        for l in range(1,rope_length):
            dist = np.sqrt((rope_pos[l-1][0]-rope_pos[l][0])**2 + (rope_pos[l-1][1]-rope_pos[l][1])**2)
        
            if dist > np.sqrt(2): # knot needs to move
                if rope_pos[l-1][0] == rope_pos[l][0]:
                    rope_pos[l][1] = int((rope_pos[l-1][1] + rope_pos[l][1]) / 2)
                elif rope_pos[l-1][1] == rope_pos[l][1]:
                    rope_pos[l][0] = int((rope_pos[l-1][0] + rope_pos[l][0]) / 2)
                else:
                    #find diagonal move:
                    d_moves = [[1,1],[1,-1],[-1,1],[-1,-1]]
                    dm_dists = [np.sqrt((rope_pos[l-1][0]-rope_pos[l][0]-d_moves[i][0])**2 + (rope_pos[l-1][1]-rope_pos[l][1]-d_moves[i][1])**2) for i in range(4)]
                    dm_move = d_moves[dm_dists.index(min(dm_dists))]
                    rope_pos[l][0] += dm_move[0]
                    rope_pos[l][1] += dm_move[1]
            if not str(rope_pos[rope_length-1]) in visited:  
                visited.append(str(rope_pos[rope_length-1]))
len(visited)
```
